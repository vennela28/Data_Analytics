```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np

# Load iris.csv into pandas DataFrame
iris = pd.read_csv("iris.csv")
```

```python
iris_setosa = iris.loc[iris["species"] == "setosa"]
iris_versicolor = iris.loc[iris["species"] == "versicolor"]
iris_virginica = iris.loc[iris["species"] == "virginica"]
```

```python
# Histogram of petal lengths of the 3 species of flowers
sns.FacetGrid(iris, hue="species", size = 5)\
    .map(sns.distplot, "petal_length") \
    .add_legend()
    
plt.show();
```

```python
# Violin plot combines the benefits of box-plots and histograms and PDF andd simplifies them. 
# Denser regions of the data are fatter and sparser ones are thinner in a violin plot.

# The thick black area is the box of box-plot where the white spot in the box is the 50%ile and have lower and upper whisker.
# The Violin plot plots the PDF on the sides.

# Violin Plots give the best of both the worlds.

sns.violinplot(x="species", y="petal_length", data=iris, size = 9)
plt.show()
```
