<h3> 3.3 Pair Plots </h3>

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np

# Load iris.csv into pandas DataFrame
iris = pd.read_csv("iris.csv")
```

```python
# pairwise scatter plot: Pair plot
# Disadvantages:
# Can be used when number of features are high.
# Cannot visualize higher dimensional patterns in 4-D. 
# Only possible to view 2D patterns.
# for 4-D data we have 4C2 = 6 plots to look at.
plt.close()
sns.set_style("whitegrid")
sns.pairplot(iris, hue="species", size = 3)
plt.show()
# Diagonal elements are PDFs for each feature.
```

<h3> Observations </h3>

- petal_length & petal_width are the most useful features to identify various flower types.
- While Setosa can be easily identified (linearly separable), Virginica and Versicolor have some overlap (almost linearly separable)
- We can find "lines" and "if-else" conditions to build a simple model to classify the flower types.

<h3> Limitations </h3>

- For higher dimensional data we will have nC2 pair plots to look at, pair plots do not help us much in visualizing data.
- So we have PDFs, Linear Algebra, Principle Component Analysis, t-SNE, Dimensionality Reduction Techniques which help us visualize higher dimensions of the data.
