```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np

# Load iris.csv into pandas DataFrame
iris = pd.read_csv("iris.csv")
```

```python
iris_setosa = iris.loc[iris["species"] == "setosa"]
iris_versicolor = iris.loc[iris["species"] == "versicolor"]
iris_virginica = iris.loc[iris["species"] == "virginica"]
```

```python
# Histogram of petal lengths of the 3 species of flowers
sns.FacetGrid(iris, hue="species", size = 5)\
    .map(sns.distplot, "petal_length") \
    .add_legend()
    
plt.show();
```

```python
# 2D Density plot : Contour Plot
sns.jointplot(x = "petal_length", y = "petal_width", data = iris_setosa, kind="kde")
plt.show()
```

<h4> Observations </h4>

- There are more points in the center and there are fewer points as you go away from the center in pair plots
- Imagine a hill coming out of the screen towards you. The height of the hill is higher in the darker region and the height of the hill is shallow around the darker region. 
- Dark blue to light blue regions are known as Contours. 
- The above plot is a Contour Probability Density Plot. 
- All the points in the same Contour have the same height. That is how the circles (contours) are created. 
- Combining 2 univariate plots on the sides gives Contour plot. 
- Give us an idea of density in a 2D space.
- 1D Density - PDF
- 2D Density - Contour PDF
