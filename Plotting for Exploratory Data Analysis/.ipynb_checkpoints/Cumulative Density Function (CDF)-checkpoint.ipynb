<h3> Cumulative Density Function </h3>

- We can visually see what % of versicolor points have a petal_length < 5
- How to construct a CDF?
- How to read a CDF?

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np

# Load iris.csv into pandas DataFrame
iris = pd.read_csv("iris.csv")
```

```python
iris_setosa = iris.loc[iris["species"] == "setosa"]
iris_versicolor = iris.loc[iris["species"] == "versicolor"]
iris_virginica = iris.loc[iris["species"] == "virginica"]

plt.plot(iris_setosa["petal_length"], np.zeros_like(iris_setosa['petal_length']))
plt.plot(iris_versicolor["petal_length"], np.zeros_like(iris_versicolor['petal_length']))
plt.plot(iris_virginica["petal_length"], np.zeros_like(iris_virginica['petal_length']))

plt.show()
```

- Blue line (PDF) and orange line (CDF)
- petal_length of Setosa flowers
- PDF represents how many points - 20% of points have petal_length > 1.5 and < 1.6
- CDF - There are 82% of Setosa flowers that have petal_length <= 1.6
- 100% of Setosa flowers have petal_length <= 1.9
<h4>CDF - What % of flowers (on y-axis ) have a value <= corresponding petal_length (x-axis point)</h4>
- All probabilities lie between 0% and 100%
- 50 Setosa data sample
- How many flowers have a petal length of < 1.6 ? 
- 41 Setosa flowers have petal_length <= 1.6. (41/50 = 82%)
<h4> Construct CDF - Calculate y-coordinate using x value </h4>
- Repeat for other points of petal_length to get the CDF curve.
                              OR
- PDF counts how many flowers are there for each petal length - plot histogram
- Sum up PDF (cumulative sum) for all the intervals before a certain interval to get CDF 

- (Calculus) The area under the curve of PDF upto that point is CDF.
- Differentiate CDF ---> PDF
- Integrate PDF ---> CDF
<h5> Plot CDF and when you build simple models ( use simple if else conditions ) - you can tell how accurate your simple model using just 1 feature could be by reading data off CDF which you cannot get from PDF. </h5>

```python
# Plot CDF of petal_length having 10 bins - SETOSA
counts, bin_edges = np.histogram(iris_setosa['petal_length'], bins = 10, density = True)
pdf = counts/(sum(counts))
print(pdf)
print(bin_edges)

# compute CDF
cdf = np.cumsum(pdf)
plt.plot(bin_edges[1:], pdf)
plt.plot(bin_edges[1:], cdf)

plt.show()
```

```python
# Plot CDF of petal_length having 20 bins - SETOSA
counts, bin_edges = np.histogram(iris_setosa['petal_length'], bins = 20, density = True)
pdf = counts/(sum(counts))
print(pdf)
print(bin_edges)

# compute CDF
cdf = np.cumsum(pdf)
plt.plot(bin_edges[1:], pdf)
plt.plot(bin_edges[1:], cdf)

plt.show()
```

<h4> Observations </h4>
- If pL < 2, then Setosa - 100%
- If pL > 2 & pL <= 5, then Virginica (y-axis points) - 95% correct, 5% wrong
- If pL > 2 & pL > 5, then Versicolor (y-axis points) - 90% correct, 10% wrong
