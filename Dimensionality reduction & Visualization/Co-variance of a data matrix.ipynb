Sᵢⱼ corresponds to iᵗʰ row & jᵗʰ column element in matrix S. 
 
Sᵢⱼ = cov(fᵢ,fⱼ). Here i: 1->d, j:1 -> d.
 
<i><u>Properties of covariance</u></i>
- cov(fᵢ,fⱼ) = cov(fⱼ,fᵢ)
- cov(fᵢ,fⱼ) = var(fᵢ)

Covariance matrix looks like all the d - diagonal elements are variances of features. 

Sᵢⱼ = Sⱼᵢ are same, S is a Symmetric matrix.

<b><i>S is a square, symmetric matrix.</i></b>

Let X be column standardized which means mean(fᵢ)ⁿᵢ₌₁ = 0 and standard deviation(fᵢ)ⁿᵢ₌₁ = 1.

The cov(f₁,f₂) = 1/n ∑ⁿᵢ₌₁(xᵢ₁ - µ₁)(xᵢ₂ - µ₂) where (xᵢ₁ - µ₁) = mean(f₁) & (xᵢ₂ - µ₂) = mean(f₂)

Because of column standardization, µ₁=0 and µ₂=0. 

So, <b><i>cov(f₁,f₂) = 1/n ∑ⁿᵢ₌₁(xᵢ₁)(xᵢ₂)</i></b>

<i>If f₁ and f₂ have been standardized <b>cov(f₁,f₂) = 1/n (f₁ᵀf₂).</b></i>
