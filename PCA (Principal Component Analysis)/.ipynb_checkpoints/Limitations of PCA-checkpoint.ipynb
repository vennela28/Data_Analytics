<h6>We want to project data from 2-D to 1-D. </h6>

- <b><i>Circle / Sphere / Hypersphere</i></b>

If we want to project on v₁, we will lose a lot of information since the dataset is like a circle. λ₁ ~ λ₂ - If we project all the data on v₁, the information / spread lost is very high. If we want to project data from 2-D to 1-D, we will lose a lot of interesting information. PCA will not do good justice since we will lose a lot of information by trying to project the data from higher dimensions D to D' where D' < D.

- <b><i>4 Clusters of data across v₁ axis and v₂ axis</i></b>

If we project all the data across v₁ axis, 2 distinct groups of points get projected on v₁ axis. Projected points are forming groups and well clustered. In the other 2 distinct groups of points when projected on v₁ axis overlap and useful information is lost. The 2 distinct groups of points which are forming good separated groups / clusters in a 2-D space are being projected in the same region.  

- <b><i>The dataset formed a Sine wave</i></b>

If we project all the data across v₁ axis, we will get a linear representation of the points (bunch of points spread uniformly) in 1-D which is a beautiful Sine wave in 2-D losing useful information.
