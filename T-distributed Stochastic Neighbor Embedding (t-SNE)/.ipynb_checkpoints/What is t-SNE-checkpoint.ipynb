<b><i>t-SNE</i></b> : It is one of the best state of art for dimensionality reduction especially for visualization of data. (invented in 2008)

<b><i>PCA</i></b> : basic, old technique

There are other techniques too like Multi Dimensional Scaling (MDS), Sammon Mapping, Graph-based techniques which have been developed in the last 20 years.

t-SNE is used in visualization technique. It is used to visualize 2-D or 3-D data which is very clean and terrific. 



<h6>t-SNE is better than PCA.

PCA tries to preserve the global shape/structure of data whereas t-SNE can try to preserve the local structure of the data.</h6>
