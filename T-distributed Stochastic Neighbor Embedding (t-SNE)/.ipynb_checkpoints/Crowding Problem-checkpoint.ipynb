<h5> Crowding Problem </h5>

t-SNE - Sometimes it is impossible to preserve distances in all the neighborhoods. 

We cannot perfectly embed these points on a 1-D space - x₁, x₂, x₃, x₄ which form a square in 2-D.

Will not be able to preserve the distances between points when embedded from 2-D space to 1-D. One of the distances 'd' will be corrupted.

<b><i>t-distribution SNE (student's t-distribution) is used to resolve the crowding problem.</i></b> - will not guarantee that the distances will be preserved but try to do the best to preserve the distances. It does not promise to solve the crowding problem since it is practically impossible to solve the crowding problem. It will try to minimize the errors we have seen above.
