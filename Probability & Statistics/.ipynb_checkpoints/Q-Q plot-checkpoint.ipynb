```python
# Q-Q plot
import numpy as np
import pylab
import scipy.stats as stats
import matplotlib.pyplot as plt

# N(0, 1)
std_normal = np.random.normal(loc= 0, scale = 1, size = 1000)

# 0 to 100th percentiles of std-normal
for i in range(0, 101):
    print(i, np.percentile(std_normal, i))
```

```python
# generate 1000 samples from N(20, 5)  - mean = 20, std_deviation = 5
# generate a plot from normal distribution
measurements = np.random.normal(loc = 20, scale = 5, size = 50000)

stats.probplot(measurements, dist = "norm", plot = plt)
pylab.show()
```

<h4> <u>Limitation</u>: If the number of observations is small, it is hard to interpret Q-Q plot. </h4>

```python
# generate a plot from uniform distribution
# x-axis : Gaussian Distribution
# y-axis : Uniform Distribution
measurements = np.random.uniform(low =-1, high = 1, size = 50000)
stats.probplot(measurements, dist = "norm", plot = plt)
pylab.show()
```

<h3> Conclusion </h3>

- If most / all of the points lie on the straight red line then we can conclude that the distribution on both x-axis and y-axis are from the same family.

<h4> 2 questions are answered from this Q-Q plot </h4>

Q-Q plot is a graphical technique to answer these 2 types of questions below.
- Is X normally distributed ~ N(mean, variance)?
- If we have random variables X, Y, do random variables X, Y have similar distribution?
