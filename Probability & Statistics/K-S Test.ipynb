# K-S Test

```python
import numpy as np
import seaborn as sns
from scipy import stats
import matplotlib.pyplot as plt

# generate a Gaussian random variable x of 1000 samples - all points aew Gaussian distributed
x = stats.norm.rvs(size=1000)
sns.set_style('whitegrid')
sns.kdeplot(np.array(x), bw=0.5)
plt.show()
```

```python
# In K-S Test, to determine the 1000 values in x follow Gaussian distribution
# pvalue is high which shows that x is normally distributed
stats.kstest(x, 'norm')
```

```python
# y - Continuous Uniform Distribution(0,1)
y = np.random.uniform(0, 1, 10000)
sns.kdeplot(np.array(y), bw=0.1)
plt.show()
```

```python
#K-S test for a uniform random distribution, p-value = 0 which means y is not normally distributed
stats.kstest(y, 'norm')   
```
